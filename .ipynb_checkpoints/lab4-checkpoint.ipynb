```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.e**(x / 3)

def df(f, x, eps):
    "metod Runge"
    h = 1
    d1 = (f(x) - f(x - h)) / h
    d2 = (f(x) - f(x - h / 2)) / h * 2 
    
    while (np.abs(d1 - d2) > eps):
        d1 = (f(x) - f(x - h)) / h
        h /= 2
        d2 = (f(x) - f(x - h)) / h
        
    print(f'{x} {eps} : h = {h}')
    return d1
    

eps1, eps2 = 10**(-3), 10**(-6)
xi = 0.9
print(f'{xi} {eps1} : df = {df(f, xi, eps1)}\n')
print(f'{xi} {eps2} : df = {df(f, xi, eps2)}')
```

```python
def d3f(f, x, eps):
    h = 1
    d1 = (d2f(f, x, eps) - d2f(f, x - h, eps)) / h
    d2 = (d2f(f, x, eps) - d2f(f, x - h / 2, eps)) / h * 2 
    
    while (np.abs(d1 - d2) > eps):
        d1 = (d2f(f, x, eps) - d2f(f, x - h, eps)) / h
        h /= 2
        d2 = (d2f(f, x, eps) - d2f(f, x - h, eps)) / h
        
    return d1

def O_2(f, x, h):
    M3 = np.abs(d3f(f, x + h, eps))
    return M3 * h**2 / 6

h = 1
print(eps)
print(O_2(f, xi, h / 2))
print(O_2(f, xi, h / 4))
print(O_2(f, xi, h / 8))

dots = []
plt.plot([0.5, 0.25, 0.125], [O_1(f, xi, h / 2), O_1(f, xi, h / 4), O_1(f, xi, h / 8)], 'r')
plt.plot([0.5, 0.25, 0.125], [O_2(f, xi, h / 2), O_2(f, xi, h / 4), O_2(f, xi, h / 8)])
plt.grid()
plt.yscale('log', base=2)
#plt.xscale('log', base=2)
```

```python
def f(x):
    return np.e**(x / 3)

def derivateOne(f, x, h):
    d1 = (f(x) - f(x - h)) / h
    return d1

def derivateTwo(f, x, h):
    d2 = (f(x + h) - f(x - h)) / (2 * h)
    return d2

def Oh(ff, h, x0, x):
    M2 = np.abs(sp.diff(sp.diff(ff, x), x).subs(x, x0+h))
    return M2 * h / math.factorial(2)

def Oh2(ff, h, x0, x):
    M3 = np.abs(sp.diff(sp.diff(sp.diff(ff, x), x), x).subs(x, x0+h))
    return M3 * h**2 / math.factorial(3)

h = [1/2, 1/4, 1/8]
x0 = 0.9
x = sp.symbols('x')
ff = np.e**(x / 3)

ohList = []
for hi in h:
    ohList.append(Oh(ff, hi, x0, x))
print(ohList)
print("")
oh2List = []
for hi in h:
    oh2List.append(Oh2(ff, hi, x0, x))
print(oh2List)

plt.plot(h, ohList, 'bo', label = 'Oh')
#plt.plot(h, oh2List, 'ro', label = 'Oh2')
plt.xlabel('h')
plt.ylabel('Oh')
plt.grid()
plt.show()
```
